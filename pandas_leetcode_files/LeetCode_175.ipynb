```python
import pandas as pd
```

```python
person = pd.DataFrame({
    'personId': [1, 2],
    'lastName': ['Wang', 'Alice'],
    'firstName': ['Allen', 'Bob']
})
```

```python
person
```

```python
address = pd.DataFrame({
    'addressId': [1, 2],
    'personId': [2, 3],
    'city': ['New York City', 'Leetcode'],
    'state': ['New York', 'California']
})
```

```python
address
```

```python
person.merge(address, how= 'left', on= 'personId')[['firstName', 'lastName', 'city', 'state']]
```
