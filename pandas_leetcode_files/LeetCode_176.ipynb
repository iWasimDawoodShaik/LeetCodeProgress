```python
import pandas as pd
```

```python
employee = pd.DataFrame({
    'id': [1, 2, 3],
    'salary': [100, 200, 300]
})
```

```python
employee
```

```python
employee['d_rnk'] = employee['salary'].rank(method= 'dense', ascending= False)
```

```python
sal = employee[employee['d_rnk'] == 2]['salary'].unique()
```

```python
pd.DataFrame({'SecondHighestSalary': sal if len(sal) != 0 else [None]})
```
