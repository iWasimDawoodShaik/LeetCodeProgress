```python
import pandas as pd
```

```python
scores = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6],
    'score': [3.5, 3.65, 4.0, 3.85, 4.0, 3.65]
})
```

```python
scores['rank'] = scores['score'].rank(method= 'dense', ascending= False)
```

```python
scores[['score', 'rank']].sort_values(by= 'score', ascending= False)
```
