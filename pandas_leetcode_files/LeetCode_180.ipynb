```python
import pandas as pd
```

```python
logs = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6, 7],
    'num': ['1', '1', '1', '2', '1', '2', '2']
})
```

```python
logs
```

```python
logs = logs.sort_values(by='id', ascending= True)
```

```python
logs['n_1'] = logs['num'].shift(-1)
```

```python
logs['n_2'] = logs['num'].shift(-2)
```

```python
logs[(logs['num'] == logs['n_1']) & (logs['n_1'] == logs['n_2'])]['num'].unique()
```

```python
pd.DataFrame({'ConsecutiveNums': logs[(logs['num'] == logs['n_1']) & (logs['n_1'] == logs['n_2'])]['num'].unique()})
```
