```python
import pandas as pd
```

```python
person = pd.DataFrame({
    'id': [1, 2, 3],
    'email': ['a@b.com', 'c@d.com', 'a@b.com']
})
```

```python
person
```

```python
df = person.groupby(by= 'email').agg(cnt = ('email', 'size')).reset_index()
```

```python
df[df['cnt'] > 1][['email']].rename(columns = {'email': 'Email'})
```
