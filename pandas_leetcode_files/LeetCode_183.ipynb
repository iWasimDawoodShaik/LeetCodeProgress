```python
import pandas as pd
```

```python
customers = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'name': ['Joe', 'Henry', 'Sam', 'Max']
})
```

```python
customers
```

```python
orders = pd.DataFrame({
    'id': [1, 2],
    'customerId': [3, 1]
})
```

```python
orders
```

```python
customers[~customers['id'].isin(orders['customerId'].unique())][['name']].rename(columns ={'name': 'Customers'})
```
