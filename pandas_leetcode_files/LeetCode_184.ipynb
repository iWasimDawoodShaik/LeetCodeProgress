```python
import pandas as pd
```

```python
employee = pd.DataFrame({
    'id' : [1, 2, 3, 4, 5],
    'name': ['Joe', 'Jim', 'Henry', 'Sam', 'Max'],
    'salary': [70000, 90000, 80000, 60000, 90000],
    'departmentId': [1, 1, 2, 2, 1]
})
```

```python
employee
```

```python
department = pd.DataFrame({
    'id': [1, 2],
    'name': ['IT', 'Sales']
})
```

```python
department
```

```python
df = employee.merge(department, how= 'inner', left_on= 'departmentId', right_on= 'id', suffixes= ('_e', '_d'))
```

```python
df['rank'] = df.groupby(by= 'departmentId')['salary'].rank(method='dense', ascending= False)
```

```python
df[df['rank'] == 1][['name_d', 'name_e', 'salary']].rename(columns= {'name_d': 'Department', 'name_e': 'Employee', 'salary': 'Salary'})
```
