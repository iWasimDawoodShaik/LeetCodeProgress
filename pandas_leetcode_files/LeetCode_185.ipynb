```python
import pandas as pd
```

```python
employee = pd.DataFrame({
    'id': [1, 2, 3, 4, 5, 6, 7],
    'name': ['Joe', 'Henry', 'Sam', 'Max', 'Janet', 'Randy', 'Will'],
    'salary': [85000, 80000, 60000, 90000, 69000, 85000, 70000],
    'departmentId': [1, 2, 2, 1, 1, 1, 1]
})
```

```python
employee
```

```python
department = pd.DataFrame({
    'id': [1, 2],
    'name': ['IT', 'Sales']
})
```

```python
department
```

```python
df = employee.merge(department, how= 'inner', left_on= 'departmentId', right_on= 'id', suffixes= ('_e', '_d'))
```

```python
df
```

```python
df['d_rnk'] = df.groupby(by= 'departmentId')['salary'].rank(method= 'dense', ascending= False)
```

```python
df
```

```python
df[df['d_rnk'] <= 3][['name_d', 'name_e', 'salary']].rename(columns= {'name_d': 'Department', 'name_e': 'Employement', 'salary': 'Salary'})
```
