```python
import pandas as pd
```

```python
weather = pd.DataFrame({
    'id': [1, 2, 3, 4],
    'recordDate': ['2015-01-01', '2015-01-02', '2015-01-03', '2015-01-04'],
    'temperature': [10, 25, 20, 30]
})
```

```python
weather
```

```python
weather['recordDate'] = pd.to_datetime(weather['recordDate'])
```

```python
weather
```

```python
weather['prevDate'] = weather['recordDate'] - pd.Timedelta(days = 1)
```

```python
weather
```

```python
df = weather.merge(weather, how= 'inner', left_on= 'prevDate', right_on= 'recordDate', suffixes= ('_c', '_p'))
```

```python
df[df['temperature_c'] > df['temperature_p']][['id_c']].rename(columns= {'id_c': 'id'})
```
