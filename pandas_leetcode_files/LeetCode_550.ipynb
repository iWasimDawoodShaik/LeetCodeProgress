```python
import pandas as pd
```

```python
activity = pd.DataFrame({
    'player_id': [1, 1, 2, 3, 3],
    'device_id': [2, 2, 3, 1, 4],
    'event_date': ['2016-03-01', '2016-03-02', '2017-06-25', '2016-03-02', '2018-07-03'],
    'games_played': [5, 6, 1, 0, 5]
})
```

```python
activity
```

```python
activity['event_date'] = pd.to_datetime(activity['event_date'])
```

```python
df = activity.groupby(by= 'player_id').agg(first_login = ('event_date', 'min')).reset_index()
```

```python
df['second_day'] = df['first_login'] + pd.Timedelta(days= 1)
```

```python
df = df.merge(activity, how= 'left', left_on= ['player_id', 'second_day'], right_on= ['player_id', 'event_date'], suffixes= ('_f', '_n'))
```
