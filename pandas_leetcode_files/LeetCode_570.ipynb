```python
import pandas as pd
```

```python
employee = pd.DataFrame({
    'id': [101, 102, 103, 104, 105, 106],
    'name': ['John', 'Dan', 'James', 'Amy', 'Anne', 'Ron'],
    'department': ['A', 'A', 'A', 'A', 'A', 'B'],
    'managerId': [None, 101, 101, 101, 101, 101]
})
```

```python
employee
```

```python
df = employee.groupby(by='managerId').agg(e_cnt = ('id', 'count')).reset_index().merge(employee, how = 'inner', left_on= 'managerId', right_on= 'id', suffixes= ('_m', '_e'))
```

```python
df[df['e_cnt'] >= 5][['name']]
```
