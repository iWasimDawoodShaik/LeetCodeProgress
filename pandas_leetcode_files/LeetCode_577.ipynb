```python
import pandas as pd
```

```python
employee = pd.DataFrame({
    'empId': [3, 1, 2, 4],
    'name': ['Brad', 'John', 'Dan', 'Thomas'],
    'supervisor': [None, 3, 3, 3],
    'salary': [4000, 1000, 2000, 4000]
})
```

```python
employee
```

```python
bonus = pd.DataFrame({
    'empId': [2, 4],
    'bonus': [500, 2000]
})
```

```python
bonus
```

```python
df = employee.merge(bonus, how= 'left', on= 'empId')
```

```python
df[(df['bonus'].isna())| (df['bonus'] < 1000)][['name', 'bonus']]
```
