```python
import pandas as pd
```

```python
insurance = pd.DataFrame({
    'pid': [1, 2, 3, 4],
    'tiv_2015': [10, 20, 10, 10],
    'tiv_2016': [5, 20, 30, 40],
    'lat': [10, 20, 20, 40],
    'lon': [10, 20, 20, 40]
})
```

```python
insurance
```

```python
insurance['lat_lon_count']= insurance.groupby(by= ['lat', 'lon'])['pid'].transform('count')
```

```python
insurance
```

```python
insurance['tiv_2015_cnt'] = insurance.groupby(by= 'tiv_2015')['pid'].transform('count')
```

```python
insurance
```

```python
pd.DataFrame({'tiv_2016': [insurance[(insurance['lat_lon_count'] == 1) & (insurance['tiv_2015_cnt'] > 1)]['tiv_2016'].sum().round(2)]})
```
